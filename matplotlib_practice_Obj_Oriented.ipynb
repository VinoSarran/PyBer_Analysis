```python
# Import dependencies.
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis,label='Boston')
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis, label = 'Boston', marker = 'd',color="green",linewidth=2)
ax.legend()
ax.grid()
ax.set_title("Yup Yup")
ax.set_xlabel('Time')
ax.set_ylabel('Value')
ax.set_ylim(0, 100)

```

```python
##Lesson 5.1.5 BAR CHARTS - object oriented
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis,y_axis,color="cyan", label='Chicago')
ax.legend()
ax.set_title("Pyber Fare by Month")
ax.set_xlabel('Fare')
ax.set_ylabel('Month')
ax.invert_yaxis()
```

```python
##SCATTER 5.1.7

fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis)
```

```python
##BUBBLE 5.1.7
fig, ax = plt.subplots()
ax.scatter(x_axis, y_axis, c="skyblue",s = [i * 100 for i in y_axis])
```

```python
## pie chart 5.1.8
```

```python
fig, ax = plt.subplots()
ax.pie(y_axis,labels=x_axis)
plt.show()
```
